<div style="text-align: Left; padding: 10px; background-color:#820b2a; color:#ffffff; font-size: 26px; box-shadow: 0px 4px 8px rgba(0, 0, 0, 0.1);">3.0 DATA PREPROCESSING</div>


As we navigate the waters of the Titanic dataset, the Data Processing phase becomes our compass, refining raw data into pearls of insight. 

Here, we cleanse, transform, and sculpt our dataset, preparing it for the predictive voyage. From handling missing values to encoding categorical features, our tools—like Python's Pandas and Scikit-Learn—become the artisans shaping the narrative. 

The OneHotEncoder was employed as a crucial component. This encoder was employed to transform the categorical variables into a numerical format, so as to facilitate the inclusion of the variables in the selected machine learning model.
Join us as we mold the data, ensuring it's a worthy vessel for the journey ahead.

```python
#Create an object of the class: OneHotEncoder

Encode_me = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
print(Encode_me)
```

```python
#Fit_Transform your variables with the Object created from the instance of the class, then assign it to a new variable

df = Encode_me.fit_transform(titanic[["Name", "Sex", "Ticket", "Cabin", "Embarked"]])
```

```python
df
```

```python
# Set the names to the original name and Conevrt it to a DataFrame 

Get_me = Encode_me.get_feature_names_out(['Name', 'Sex', 'Ticket', 'Cabin', 'Embarked'])

onehot_df = pd.DataFrame(df, columns = Get_me)
```

```python
onehot_df
```

```python
#Set your original DataSet to the Encoded DataSet by dropping the previous variables in their formal forms and assigning it to the new variables in
#their original form

titanic = pd.concat([titanic, onehot_df], axis = 1).drop(columns = ["Name", "Sex", "Ticket", "Cabin", "Embarked"])
```

### 3.1 Splitting the dataset into train and test sets

In our Titanic Prediction System, we navigate the data divide by splitting it into two essential realms: the Training Set and the Test Set. 

The Training Set becomes the proving ground for our predictive models, while the Test Set remains untouched, ready to evaluate their seaworthiness.

In this case, 80% of the dataset was kept for training and the remaining 20% for splitting.

With this division, we ensure our models set sail prepared, having learned from the past and ready to face the challenges of the uncharted future. 

```python
titanic["Survived"]
```

```python
y = titanic["Survived"]
X = titanic.drop("Survived", axis = 1)
```

```python
trainX, testX, trainy, testy = train_test_split(X, y, test_size=0.2)
```

## 4.1 DECISION TREE CLASSIFIER

In our Titanic Prediction System, the Decision Tree Classifier stands as a key player in our ensemble of predictive models. 

This algorithm, like the branching paths of a tree, makes decisions based on features, guiding us through the intricacies of survival prediction.

As we traverse this data woodland, each split becomes a clue, and each leaf a revelation. 

 

```python
model1 = DecisionTreeClassifier()
```

```python
model1.fit(trainX, trainy)
pred1 = model1.predict(testX)
pred1
```

```python
print(classification_report(pred1,testy))
```

```python
report1 = classification_report(testy, pred1, target_names = ['class 0', 'class 1'])

recall = recall_score(pred1, testy)
precision = precision_score(pred1, testy)
fi_score = f1_score(pred1, testy)

performance_table.update({'Decision Tree': [recall, precision, fi_score]})
performance_table
```

## 4.2 LOGISTIC REGRESSION

In the seas of predictive modeling for our Titanic Prediction System, Logistic Regression serves as our compass. 

Unlike its linear counterpart, this algorithm charts the probability waves, determining the likelihood of survival. 

As we voyage through numerical waters, Logistic Regression becomes our guide, predicting outcomes in a world that's not black and white but steeped in probabilities. 

```python
logistic_model = LogisticRegression()
```

```python
logistic_model.fit(trainX, trainy)
```

```python
pred_logistic = logistic_model.predict(testX)
pred_logistic
```

```python
print(classification_report(pred_logistic,testy))
```

```python
report2 = classification_report(testy,pred_logistic, target_names = ['class 0', 'class 1'])

recall = recall_score(pred_logistic, testy, average='micro')
precision = precision_score(pred_logistic, testy, average='micro')
fi_score = f1_score(pred_logistic, testy, average='micro')

performance_table.update({'Logistic Regression': [recall, precision, fi_score]})

performance_table
```

## 4.3 KNEIGHBORS CLASSIFIER 

Picture the k-Neighbors Classifier as a friendly guide in our Titanic Prediction System. It makes predictions based on what its closest neighbors think. If they lean towards survival or otherwise, so does our prediction. It's like making decisions by checking with nearby friends. 

```python
New = KNeighborsClassifier()
```

```python
New.fit(trainX, trainy)
```

```python
Predict_Me = New.predict(testX)
```

```python
print(classification_report(Predict_Me,testy))
```

```python
report3 = classification_report(testy,Predict_Me, target_names = ['class 0', 'class 1'])

recall = recall_score(Predict_Me, testy, average='micro')
precision = precision_score(Predict_Me, testy, average='micro')
fi_score = f1_score(Predict_Me, testy, average='micro')

performance_table.update({'KNN': [recall, precision, fi_score]})
```

## 4.4 RANDOM FOREST CLASSIFIER

In the realm of our Titanic Prediction System, the Random Forest Classifier is like a wise forest, where each tree holds a unique perspective. 

It combines the collective wisdom of many trees, each contributing to a decision based on its own experiences. 

Together, they form a resilient forest that navigates the complexities of survival prediction. Join us as we venture into this algorithmic woodland, where the strength lies in the diversity of perspectives.

```python
Random = RandomForestClassifier()
```

```python
Random.fit(trainX, trainy)
```

```python
Random_Predict = Random.predict(testX)
```

```python
print(classification_report(Random_Predict,testy))
```

```python
report4 = classification_report(testy,Random_Predict, target_names = ['class 0', 'class 1'])

recall = recall_score(Random_Predict, testy, average='micro')
precision = precision_score(Random_Predict, testy, average='micro')
fi_score = f1_score(Random_Predict, testy, average='micro')

performance_table.update({'Random Forest': [recall, precision, fi_score]})
```

```python
performance_table = pd.DataFrame(performance_table, index=['recall', 'precision', 'fi_score'])

performance_table = performance_table.T

```
